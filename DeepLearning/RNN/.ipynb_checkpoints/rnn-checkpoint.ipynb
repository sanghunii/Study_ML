```python
##IMDB dataset을 이용해서 긍정리뷰/부정리뷰 분류해보기
##Prepare dataset (train, validation, test)

from tensorflow.keras.datasets import imdb
(train_input, train_target), (test_input, test_target) = imdb.load_data(
    num_words=300)
print(train_input.shape, train_target.shape)

print(len(train_input[0]), len(train_input[1]))
print('##1번째 문장에 대한 indexing이 어떻게 이루어져 있는지 알아보자 ##\n', train_input[:1])
print(train_target[:20])

from sklearn.model_selection import train_test_split
train_input, val_input, train_target, val_target = train_test_split(
    train_input, train_target, test_size = 0.2, random_state = 42)

print(train_input.shape, val_input.shape, test_input.shape)

import numpy as np
lengths = np.array([len(x) for x in train_input])    #이런식으로 배열 생성하는 코드들 유심하게 봐놓고 이용하기
print(np.mean(lengths), np.median(lengths))

import matplotlib.pyplot as plt
plt.figure()
plt.title('Length of IMDB Review Data')
plt.hist(lengths)
plt.xlabel('length')
plt.ylabel('frequency')
plt.show()
plt.close()

from tensorflow.keras.preprocessing.sequence import pad_sequences
train_seq = pad_sequences(train_input, maxlen=100)
#maxlen=100으로 지정함에 따라서 토큰의 길이가 100이 넘는 리뷰데이터는 앞부분이 짤릴까? 뒷부분이 짤릴까? 아래코드에서 알아보자

print(train_input.shape)  #파이썬 리스트들의 배열 즉 1-dimension array
print(train_seq.shape) #2차원 배열    차이점을 잘 알고 있어야 한다.
print(train_seq[0])
print(train_input[0][-10:])

print(train_seq[5])
print(np.array(train_input[5]).shape)

val_seq = pad_sequences(val_input, maxlen=100)
print(val_seq.shape)
```

```python
#Make Recurrent Neural Network - One-Hoe-Encoding
from tensorflow import keras
model = keras.Sequential()
model.add(keras.layers.SimpleRNN(8, input_shape=(100, 300)))
model.add(keras.layers.Dense(1, activation='sigmoid'))

#Match the each input token to one-hot-incoding / use keras.utils.to_categorical
train_oh = keras.utils.to_categorical(train_seq)
print(train_seq.shape)
print(train_oh.shape)
print(train_oh[0][0][:12])
print(train_oh[0][0].sum())
val_oh = keras.utils.to_categorical(val_seq)

model.summary()
#Sequence()이용해서 모델 구조 완성했고
#train, validation dataset에 대한 preprocessing도 완료했다.
#이제 model.compile()을 이용해서 해당 모델에서 사용할 optimizer, lossfunction등을 지정하고 fit()을 이용해서 training하자.

rmsprop = keras.optimizers.RMSprop(learning_rate=1e-4) #1e-4 = 0.0001 e-4 = 10의 -4제곱
model.compile(
    optimizer=rmsprop,
    loss='binary_crossentropy',
    metrics=['accuracy']
)
checkpoint_cb = keras.callbacks.ModelCheckpoint('best-simplernn-model.keras',
                                                save_best_only = True)
early_stopping_cb = keras.callbacks.EarlyStopping(patience=3,
                                                  restore_best_weights=True)
history = model.fit(train_oh, train_target, epochs=100, batch_size=64,
                    validation_data=(val_oh, val_target),
                    callbacks=[checkpoint_cb, early_stopping_cb])
```

```python
plt.figure()
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('improvement log of RNN model_loss')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['train','validation'])
plt.show()

plt.figure()
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('improvement log of RNN model_accuracy')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['train','validation'])
plt.show()

##단, 이렇게 one-hot-encoding을 통해서 어휘 사전의 각 단어들(token들)을 labeling하게 되면 입력의 크기가 매우 커진다.
print(train_seq.nbytes, train_oh.nbytes)
print(train_seq.dtype)
print(train_oh.dtype)
```

```python
###Using word embedding!!!!!!!!!
model2 = keras.Sequential()
model2.add(keras.layers.Embedding(300, 16, input_length=100, input_shape=(100,)))
model2.add(keras.layers.SimpleRNN(8))
model2.add(keras.layers.Dense(1, activation='sigmoid'))
#model2.build(input_shape=(None, 100))
model2.summary()

rmsprop = keras.optimizers.RMSprop(learning_rate=1e-4) #1e-4 = 0.0001 e-4 = 10의 -4제곱
model2.compile(
    optimizer=rmsprop,
    loss='binary_crossentropy',
    metrics=['accuracy']
)
checkpoint_cb = keras.callbacks.ModelCheckpoint('best-embedding-model.keras',
                                                save_best_only = True)
early_stopping_cb = keras.callbacks.EarlyStopping(patience=3,
                                                  restore_best_weights=True)
history = model2.fit(train_seq, train_target, epochs=100, batch_size=64,
                    validation_data=(val_seq, val_target),
                    callbacks=[checkpoint_cb, early_stopping_cb])
```

```python

plt.figure()
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('improvement log of RNN model_loss')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['train','validation'])
plt.show()

plt.figure()
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('improvement log of RNN model_accuracy')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['train','validation'])
plt.show()
```
