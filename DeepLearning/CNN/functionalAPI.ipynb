```python
#####2절에서 best-cnn-model.keras파일을 생성한 다음 실행해야한다.######
from tensorflow import keras
model = keras.models.load_model('best-cnn-model.keras')
model.summary()
model.layers


#print('input : ', model.input)
#print('Inputs: ', model.inputs)

conv = model.layers[0]
print(conv.weights[0].shape, conv.weights[1].shape)

conv_weights = conv.weights[0].numpy()
print(conv_weights.mean(), conv_weights.std())

import matplotlib.pyplot as plt
plt.figure()
plt.title('Histogram of Weight') ##Q) Histogram과 Bar의 다른점에 대해서 서술 : 
plt.hist(conv_weights.reshape(-1, 1))
plt.xlabel('weight')
plt.ylabel('count')
plt.show()

#32개의 커널을 16개씩 두 줄에 걸쳐서 출력해 보자
fig, axs = plt.subplots(2, 16, figsize=(15,2))
for i in range(2):
    for j in range(16):
        axs[i,j].imshow(conv_weights[:,:,0,i*16+j], vmin=-0.5, vmax=0.5)
        axs[i,j].axis('off')
plt.show()
```

```python
#훈련하지 않은 모델의 filter를 그려서 비교해보자
no_training_model = keras.Sequential()
no_training_model.add(keras.layers.Conv2D(32,
                                          kernel_size=3,
                                          activation='relu',
                                          padding='same',
                                          input_shape=(28,28,1)))
no_training_conv = no_training_model.layers[0]
print(no_training_conv.weights[0].shape)

no_training_weights = no_training_conv.weights[0].numpy()
print(no_training_weights.mean(), no_training_weights.std())

plt.figure()
plt.title('No_Training')
plt.hist(no_training_weights.reshape(-1,1))
plt.xlabel('weight')
plt.ylabel('count')
plt.show()

print('reshape structure : ', no_training_weights.reshape(-1,1).shape)
print('origin shape : ', no_training_weights.shape)

fig, axs = plt.subplots(2, 16, figsize=(15,2))
for i in range(2):
    for j in range(16):
        axs[i,j].imshow(no_training_weights[:,:,0,i*16+j], vmin=-0.5, vmax=0.5)
        axs[i,j].axis('off')
plt.show()
```

```python
#keras functional API- Visualize feature map
#Functional API이용해서 일단 모델 구현 먼저 해보자
#Check 1st convolution layer feature map
print(model.inputs[0])   #책에는 model.inputs가 아니라 model.input이라고 나와있다, 고쳐야 할듯?
conv_acti = keras.Model(model.inputs[0], model.layers[0].output)

(train_input, train_target), (test_input, test_target) =\
    keras.datasets.fashion_mnist.load_data()

plt.imshow(train_input[0], cmap='gray_r')
plt.show()

inputs = train_input[0:1].reshape(-1,28,28,1) / 255.0
feature_maps = conv_acti.predict(inputs)
print('##1', feature_maps.shape)

fig, axs = plt.subplots(4, 8, figsize=(15,8))
for i in range(4):
    for j in range(8):
        axs[i,j].imshow(feature_maps[0, :, :, i*8 + j])
        axs[i,j].axis('off')
plt.show()


#Check 2nd convolution layer feature map
conv2_acti = keras.Model(model.inputs[0], model.layers[2].output)
inputs = train_input[0:1].reshape(-1, 28, 28, 1) / 255.0
print(inputs.shape)
feature_maps = conv2_acti.predict(inputs)
print(feature_maps.shape)
print('###### : ', feature_maps[0, :, :, 1].shape)


fig, axs = plt.subplots(8, 8, figsize=(12,12))
for i in range(8):
    for j in range(8):
        axs[i,j].imshow(feature_maps[0, :, :, i*8 + j])
        axs[i,j].axis('off')
plt.show()
```
