```python
##Neural Network

from tensorflow import keras
(train_input, train_target), (test_input, test_target) =\
    keras.datasets.fashion_mnist.load_data()

print(train_input.shape, train_target.shape)
print(test_input.shape, test_target.shape)

import matplotlib.pyplot as plt
plt.figure()
fig, axs = plt.subplots(1, 10, figsize=(10,10))
for i in range(10):
    axs[i].imshow(train_input[i], cmap = 'gray_r')
    axs[i].axis('off')
plt.show()

import numpy as np
print([train_target[i] for i in range(10)])
print(np.unique(train_target, return_counts = True))
```

```python
###First we use Logistic
train_scaled = train_input/255.0
train_scaled = train_scaled.reshape(-1, 28*28)
print(train_scaled.shape)

from sklearn.model_selection import cross_validate
from sklearn.linear_model import SGDClassifier
sc = SGDClassifier(loss='log_loss', max_iter=5, random_state=42)
scores = cross_validate(sc, train_scaled, train_target, return_train_score = True, n_jobs=-1)
print('test(validate) scores : ', np.mean(scores['test_score']))
print('train_scores : ', np.mean(scores['train_score']))
print(scores)
```

```python
##show test_score changes
max_iteration = [5, 10, 100, 500, 10000]
test_scores = []
train_scores = []
for i in max_iteration:
    sc = SGDClassifier(loss= 'log_loss', max_iter = i, random_state=42)
    scores = cross_validate(sc, train_scaled, train_target, return_train_score = True, n_jobs=-1)
    test_scores.append(np.mean(scores['test_score']))
    train_scores.append(np.mean(scores['train_score']))
print('each test score : ', test_scores)
print('each train score : ', train_scores)
plt.figure()
plt.title('score_changes')
plt.plot(max_iteration, test_scores, label = 'test_socre')
plt.plot(max_iteration, train_scores, label = 'train_score')
plt.legend()
plt.show()
plt.close()
##As we can see on the plot, there are limits to the improvement of train_score and test_score.
```

```python
####Use tensorflow - keras
from tensorflow import keras

from sklearn.model_selection import train_test_split
train_scaled, val_scaled, train_target, val_target = train_test_split(
    train_scaled, train_target, test_size = 0.2, random_state = 42)
print(train_scaled.shape, train_target.shape)
print(val_scaled.shape, val_target.shape)

#make artificial neural network model
dense = keras.layers.Dense(10, activation = 'softmax', input_shape=(784,))
model = keras.Sequential([dense])

model.compile(
    loss='sparse_categorical_crossentropy', 
    metrics=['accuracy']
)

model.fit(train_scaled, train_target, epochs=5)
model.evaluate(val_scaled, val_target)
```
