```python
##Data preprocessing - standard score (different with standard deviation)
#We make machine learning model that predict weight of fish when given length

fish_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0,
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0,
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0, 9.8,
                10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
fish_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0,
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0,
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0, 6.7,
                7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]

import numpy as np

#numpy    input data 
fish_data = ((fish_length, fish_weight))
print(fish_data)
fish_data = np.column_stack((fish_length, fish_weight))
print(fish_data)

#numpy    target data 
#1 -> positive, 2 -> negative
fish_target = np.concatenate((np.ones(35), np.zeros(14)))

from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(
    fish_data, fish_target, random_state = 42)

print(train_input.shape, test_input.shape)
print(train_target.shape, test_target.shape)

print('#1 : ', test_target)

train_input, test_input, train_target, test_target = train_test_split(
    fish_data, fish_target, stratify = fish_target, random_state = 42)
#Stratify parameter is used for maintain proportion of class when dispart dataset (In this case, we break fish_data into train_input, tset_input)
#It is useful when classes are imbalanced distribution
print('#2 : ', test_target)


from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()                   
kn.fit(train_input, train_target)             
print(kn.score(test_input, test_target))
print(kn.predict([[25, 150]]))                


import matplotlib.pyplot as plt
plt.figure()
plt.title('#1 - initial plot')
plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(25, 150, marker='^')               
plt.xlabel("length")
plt.ylabel('weight')
plt.show()                                     

distances, indexes = kn.kneighbors([[25, 150]])


plt.figure()
plt.title('#2 - find cause')
plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(25, 150, marker='^')
plt.scatter(train_input[indexes,0], train_input[indexes, 1], marker = 'D')  # indexes . (     )
plt.xlabel("length")
plt.ylabel('weight')
plt.show()

print(train_input[indexes])
print(train_target[indexes])
print(distances)               
#    x   y  . -->    .
#         scale   .

plt.figure()
plt.title('#3 - make x\'s scale same with y')
plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(25, 150, marker='^')
plt.scatter(train_input[indexes,0], train_input[indexes, 1], marker='D')
plt.xlim((0, 1000))
plt.xlabel('length')
plt.ylabel('weight')
plt.show()



##Apply standard scroe
#       . --> data preprocessing( )
#      standard score(== z score)
# :         .   -->How many times the standard deviation is each feature value from the mean?
mean = np.mean(train_input, axis = 0) #  (axis = 0     2 list  length weight  )
print('mean : ', mean)
std = np.std(train_input, axis = 0)   #   (  )

train_scaled = (train_input - mean) / std   #braodcasting
print(train_scaled)


plt.figure()
plt.title('Before apply to sample')
plt.scatter(train_scaled[:,0], train_scaled[:,1])
plt.scatter(25, 150, marker='^')
plt.xlabel("length")
plt.ylabel('weight')
plt.show()                

plt.figure()
plt.title('After apply to sample')
new = ([25, 150] - mean) / std    #turn new value into standard score value
plt.scatter(train_scaled[:,0], train_scaled[:,1])
plt.scatter(new[0], new[1], marker='^')
plt.xlabel("length")
plt.ylabel('weight')
plt.show()               



kn.fit(train_scaled, train_target)
test_scaled = (test_input - mean) / std
#generate 'broadcasting' in above code
print(kn.predict([new]))
distances, indexes = kn.kneighbors([new])
print('distances : ', distances)
print('5-nearest value : \n', train_input[indexes])
plt.figure()
plt.title('Conclusion')
plt.scatter(train_scaled[:,0], train_scaled[:,1]) 
plt.scatter(new[0], new[1], marker='^')
plt.scatter(train_scaled[indexes,0], train_scaled[indexes, 1], marker='D')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```
