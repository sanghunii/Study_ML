```python
#multiple_regression
#reidge_regression
#lasso regression

#multiple_regression
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/rickiepark/hg-mldl/master/perch_full.csv')
perch_full = df.to_numpy()
print(perch_full)

import numpy as np
perch_length = np.array([8.4, 13.7, 15.0, 16.2, 17.4, 18.0, 18.7, 19.0, 19.6, 20.0, 21.0,
       21.0, 21.0, 21.3, 22.0, 22.0, 22.0, 22.0, 22.0, 22.5, 22.5, 22.7,
       23.0, 23.5, 24.0, 24.0, 24.6, 25.0, 25.6, 26.5, 27.3, 27.5, 27.5,
       27.5, 28.0, 28.7, 30.0, 32.8, 34.5, 35.0, 36.5, 36.0, 37.0, 37.0,
       39.0, 39.0, 39.0, 40.0, 40.0, 40.0, 40.0, 42.0, 43.0, 43.0, 43.5,
       44.0])
perch_weight = np.array([5.9, 32.0, 40.0, 51.5, 70.0, 100.0, 78.0, 80.0, 85.0, 85.0, 110.0,
       115.0, 125.0, 130.0, 120.0, 120.0, 130.0, 135.0, 110.0, 130.0,
       150.0, 145.0, 150.0, 170.0, 225.0, 145.0, 188.0, 180.0, 197.0,
       218.0, 300.0, 260.0, 265.0, 250.0, 250.0, 300.0, 320.0, 514.0,
       556.0, 840.0, 685.0, 700.0, 700.0, 690.0, 900.0, 650.0, 820.0,
       850.0, 900.0, 1015.0, 820.0, 1100.0, 1000.0, 1100.0, 1000.0,
       1000.0])

from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(perch_full, perch_weight, random_state = 42)


from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures()
poly.fit([[2,3]])
print(poly.transform([[2,3]]))

poly = PolynomialFeatures(include_bias = False)
poly.fit([[2,3]])
print(poly.transform([[2,3]]))

print(train_input.shape)
poly = PolynomialFeatures(include_bias = False)
poly.fit(train_input)
train_poly = poly.transform(train_input)
print(train_poly.shape)
print(poly.get_feature_names_out())
test_poly = poly.transform(test_input)
# poly.fit(test_input)   train_input set fit() poly transform()  test_input .
#  train set  test set     .


from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(train_poly, train_target)
print('train_set_score : ', lr.score(train_poly, train_target))
print('test_set_score : ', lr.score(test_poly, test_target))


#How about using higher order-term like 5-degree? --> model would be complexer --> train_set_score may be up --> But risk of overfitting is up too
poly = PolynomialFeatures(degree = 5, include_bias = False)
poly.fit(train_input)
train_poly = poly.transform(train_input)
test_poly = poly.transform(test_input)
print(poly.get_feature_names_out())
print(train_poly.shape)


lr.fit(train_poly, train_target)
print('train_set_score : ', lr.score(train_poly, train_target))
print('test_set_score : ', lr.score(test_poly, test_target))
#test_set_score is minus.. why? 
#now let's learn about 'regularization' that use for avoid overfitting instead of reduce dimention of data(same with reduce number of feature)


#first of all we have to make data standardization
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(train_poly)
train_scaled = ss.transform(train_poly)
test_scaled = ss.transform(test_poly)



#there is two model called ridge and lasso model that add regularization to linear regression model
#Ridge applies regularization based on the coefficient squared.
#Lasso apllies regularization based on the abolute value of coefficient


#Ridge regression
print('\n\nFrom now, contents about Ridge regression\n')
from sklearn.linear_model import Ridge
ridge = Ridge()
ridge.fit(train_scaled, train_target)
print(ridge.score(train_scaled, train_target))
print(ridge.score(test_scaled, test_target))

#We can use R^2 to find optical alpha(hyper parameter)value
import matplotlib.pyplot as plt
#make two empty lists
train_score = []
test_score = []

#Now, let's train the model by increasing the alpha value by 10 times from 0.001 to 100 
#and save the train_set score and test_set score of each training result in a pre-made list.
alpha_list = [0.001, 0.01, 0.1, 1, 10, 100]
for alpha in alpha_list:
    #make ridge model
    ridge = Ridge(alpha = alpha)
    #train model
    ridge.fit(train_scaled, train_target)
    train_score.append(ridge.score(train_scaled, train_target))
    test_score.append(ridge.score(test_scaled, test_target))
    
#Now, show the data using matplotlib
#We can label each graph using label parameter
plt.figure()
#We can label each graph using label parameter
plt.plot(np.log10(alpha_list), train_score, label = 'train_score')
plt.plot(np.log10(alpha_list), test_score, label = 'test_score')
#And add legend
plt.legend()
plt.xlabel('alpha')
plt.ylabel('R^2')
plt.show()
#As we can in plot, when alpha = 10^(-1), we can get optimal answer
#We can choose value of alpha 10^(-1). Let's train again using optimal alpha value
ridge = Ridge(alpha = 0.1)
ridge.fit(train_scaled, train_target)
print('train_score(ridge model, alpha = 0.1) : ', ridge.score(train_scaled, train_target))
print('test_score(ridge model, alpha = 0.1) : ', ridge.score(test_scaled, test_target))
print('\n\n\n')



#Lasso regression
from sklearn.linear_model import Lasso
lasso = Lasso()
lasso.fit(train_scaled, train_target)
print(lasso.score(train_scaled, train_target))

train_score = []
test_score = []
alpha_list = [0.001, 0.01, 0.1, 1, 10, 100]
for alpha in alpha_list:
    #make lasso model
    lasso = Lasso(alpha = alpha, max_iter = 10000)
    #train lasso model
    lasso.fit(train_scaled, train_target)
    #save each score about train_set and test_set
    train_score.append(lasso.score(train_scaled, train_target))
    test_score.append(lasso.score(test_scaled, test_target))

#Show the data 
plt.figure()
plt.plot(np.log10(alpha_list), train_score, label = 'train_score')
plt.plot(np.log10(alpha_list), test_score, label = 'test_score')
plt.legend()
plt.xlabel('alpha')
plt.ylabel('R^2')
plt.show()

#Optimal alpha value is 10. Let's train again lasso model using alpha = 10
lasso = Lasso(alpha = 10)
lasso.fit(train_scaled, train_target)
print('train score(lasso model, alpha = 10) : ', lasso.score(train_scaled, train_target))
print('test score(lasso model, alpha = 10) : ', lasso.score(test_scaled, test_target))

#Lasso model can make coefficient value 0
#Let's count how many model make 0
print(np.sum(lasso.coef_ == 0))
#In Lasso, coefficient of feature is 0 means that feature is meaningless
#So Lasso model is useful when there is too many meaningless features or we have to choose the features

#As a result, ridge model has better performance in our problem
```
